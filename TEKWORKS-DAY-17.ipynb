This suggests that customers tend to tip more, on average, when riding in yellow taxis than in green taxis.

### Hypothesis:
- H0:distribution of tip for yello=distribution of tip for green
- H1:distribution of tip for yellow != distribution of tip for green

```python
yellow_tip=df['tip'][df['color']=='yellow']
green_tip=df['tip'][df['color']=='green']
```

```python
yellow_tip
```

```python
print(green_tip.shape)
print(yellow_tip.shape)
```

```python
# Independent T-Test
from scipy import stats


t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

```

```python
alpha=0.05  # significance level
if p_val<alpha:
    print("Reject null hypothesis : The tip for yellow category is different that of green")
else:
    print("Fail to reject null hypothesis")
    
```

### Hypothesis:

```python
credit_tip=df['tip'][df['payment']=='credit card']
cash_tip=df['tip'][df['payment']=='cash']
```

**Hypothesis**
- H0 : Distribution of tip is the same across all pickup_borough categories
- H1 : At least one category has a different distribution of tip

```python
# ANOVA Test
manhattan_tip = df['tip'][df['pickup_borough']=='Manhattan']
queens_tip = df['tip'][df['pickup_borough']=='Queens']
bronx_tip = df['tip'][df['pickup_borough']=='Bronx']
brooklyn_tip = df['tip'][df['pickup_borough']=='Brooklyn']    
                
f_stat, p_val = stats.f_oneway(manhattan_tip,queens_tip,bronx_tip,brooklyn_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Spending Score differs by Region")
else:
    print("Fail to reject null hypothesis: No significant difference by Region")

```

**Hypothesis for Turkey test(for each pair pf groups):**
- H0 : distribution of group1 == distribution of group2
- H1 : distribution of group1 != distribution of group2

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd



# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```

### Hypothesis for  correlation for numeric variables:
- H0 : correlation is zero
- H1 : there is significant correlation

```python
r,p_val=stats.spearmanr(df['tip'],df['passengers'])
print("correlation ",r)
print("P-value=",p_val)
if p_val<0.05:
    print("Reject null hypothesis:There is no relationship ")
else:
    print("Fail to reject null hypothesis:There is significant relationship")
    
```
